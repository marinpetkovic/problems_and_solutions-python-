# Sub-string divisibility
Problem 43

The number, 1406357289, is a 0 to 9 pandigital number because it is made up of each of the digits 0 to 9 in some order, but it also has a rather interesting sub-string divisibility property.

Let d1 be the 1st digit, d2 be the 2nd digit, and so on. In this way, we note the following:

* d2d3d4=406 is divisible by 2
* d3d4d5=063 is divisible by 3
* d4d5d6=635 is divisible by 5
* d5d6d7=357 is divisible by 7
* d6d7d8=572 is divisible by 11
* d7d8d9=728 is divisible by 13
* d8d9d10=289 is divisible by 17

Find the sum of all 0 to 9 pandigital numbers with this property.

```python
import itertools
primes = [2,3,5,7,11,13,17]
#s is the sum we are looking for
s = 0

#loop through all the permutations that do not start with '0'
for x in itertools.permutations('0123456789'):
  has_property = True
  if x[0] == '0':
    continue

  #check the divisibility of the 7 numbers:
  for i in range(7):
    num = ""
    #num is 0 if the given 3 digit number is divisible by the i'th prime
    num = int(num.join(x[1+i:4+i]))%primes[i]
    if num != 0:
      has_property = False
      break
  #continue to next permutation if it doesn't have the property    
  if has_property == False:
    continue
  #add the number to s
  num = ""
  s += int(num.join(x))
print(s)
```

# 42. Trapping Rain Water

Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

 

**Example 1:**


Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]

Output: 6

Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.


**Example 2:**


Input: height = [4,2,0,3,2,5]

Output: 9

**Constraints:**

n == height.length

1 <= n <= 2 * 104

0 <= height[i] <= 105

```python
class Solution(object):
    def trap(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        n = len(height)

        #highestleft/right[i] is the highest elevation to the left/right of position i (including i)
        #the volume of water over position i will be equal to min(highestright[i],highestleft[i])-height[i]
        
        highestleft = [0] * n
        highestright = [0] * n
        
        for i,h in enumerate(height):
            #if i==0, there is nothing to do for highestleft
            if i == 0:
                highestleft[i] = h
            #otherwise highestleft is max of height[i] and all the previous heights, which is already computed in highestleft[i-1]
            else:    
                highestleft[i] = max(h, highestleft[i-1])
        #next we do the same in reverse to compute heighestright
        for i in range(n):
            if i == 0:
                highestright[n-1-i] = height[n-1-i]
            else:
                highestright[n-1-i] = max(height[n-1-i], highestright[n-i])
        sol = 0

        #add
        for i in range(n):
            sol += min(highestright[i], highestleft[i]) - height[i]
        
        return sol

s = Solution()

print(s.trap([0,1,0,2,1,0,1,3,2,1,2,1]))
print(s.trap([4,2,0,3,2,5]))
```

# 1406. Stone Game III


Alice and Bob continue their games with piles of stones. There are several stones arranged in a row, and each stone has an associated value which is an integer given in the array stoneValue.

Alice and Bob take turns, with Alice starting first. On each player's turn, that player can take 1, 2, or 3 stones from the first remaining stones in the row.

The score of each player is the sum of the values of the stones taken. The score of each player is 0 initially.

The objective of the game is to end with the highest score, and the winner is the player with the highest score and there could be a tie. The game continues until all the stones have been taken.

Assume Alice and Bob play optimally.

Return "Alice" if Alice will win, "Bob" if Bob will win, or "Tie" if they will end the game with the same score.

**Example 1**

Input: values = [1,2,3,7]

Output: "Bob"

Explanation: Alice will always lose. Her best move will be to take three piles and the score become 6. Now the score of Bob is 7 and Bob wins.

**Example 2.**

Input: values = [1,2,3,-9]

Output: "Alice"

Explanation: Alice must choose all the three piles at the first move to win and leave Bob with negative score.
If Alice chooses one pile her score will be 1 and the next move Bob's score becomes 5. In the next move, Alice will take the pile with value = -9 and lose.
If Alice chooses two piles her score will be 3 and the next move Bob's score becomes 3. In the next move, Alice will take the pile with value = -9 and also lose.
Remember that both play optimally so here Alice will choose the scenario that makes her win.


**Constraints:**

1 <= stoneValue.length <= 5 * 104

-1000 <= stoneValue[i] <= 1000

```python
class Solution:
    def stoneGameIII(self, stoneValue) -> str:
                
        n = len(stoneValue)
        
        #we use dynammic programming
        #dp[i] is the maximal advantage Alice can have after restarting the game with i rocks already taken,
        #then the winner will be given by the sign of dp[0]

        dp = [None]*n
        
        #if all but one rocks are taken, Alice wins the value of the last stone, and Bob wins nothing
        dp[-1] = stoneValue[-1]

        #take care of the edge case if n==1
        if n == 1:
            if dp[0] > 0:
              #dp[0] > 0 means Alice wins more than Bob
                return "Alice"
            elif dp[0] == 0:
                return "Tie"
            else:
                return "Bob"

        #if two rocks are remaining, Alice can either take one stone and Bob the other, or Alice takes both        
        dp[-2] = max(stoneValue[-2]+dp[-1],stoneValue[-2]-dp[-1])

        #another edge case before we can write the recursion
        if n == 2:
            if dp[0] > 0:
                return "Alice"
            elif dp[0] == 0:
                return "Tie"
            else:
                return "Bob"

        #If there are 3 stones left, Alice chosses the maximal option of 1,2 or 3 rocks minus maximal strategy for Bob with the remaining rocks (if any)  
        dp[-3] = max([stoneValue[-3] - dp[-2], stoneValue[-3] + stoneValue[-2] - dp[-1], stoneValue[-3] + stoneValue[-2] + stoneValue[-1]])
        
        #compute the sequence recursively in reverse
        for i in reversed(range(0, n-3)):
            dp[i] = max([-dp[i+1] + stoneValue[i], -dp[i + 2] + sum(stoneValue[i: i + 2]), -dp[i + 3] + sum(stoneValue[i: i + 3])])
        
        if dp[0] > 0:
            return "Alice"
        elif dp[0] == 0:
            return "Tie"
        else:
            return "Bob"

Solution().stoneGameIII([1,2,3,7])
```

# 1510. Stone Game IV

Alice and Bob take turns playing a game, with Alice starting first.

Initially, there are n stones in a pile. On each player's turn, that player makes a move consisting of removing any non-zero square number of stones in the pile.

Also, if a player cannot make a move, he/she loses the game.

Given a positive integer n, return true if and only if Alice wins the game otherwise return false, assuming both players play optimally.

**Example 1:**

Input: n = 1

Output: true

Explanation: Alice can remove 1 stone winning the game because Bob doesn't have any moves.

**Example 2:**

Input: n = 2

Output: false

Explanation: Alice can only remove 1 stone, after that Bob removes the last one winning the game (2 -> 1 -> 0).

**Example 3:**

Input: n = 4

Output: true

Explanation: n is already a perfect square, Alice can win with one move, removing 4 stones (4 -> 0).

```python
class Solution:
    def winnerSquareGame(self, n: int) -> bool:
        
        
    
        win = [False]*(n + 1)

        #win[i] is true if i remaining stones is a winning position for Alice
        #the answer will be given in win[n]



        #if i is a square, win[i] is True, because the player can remove all the stones and win
        for i in range(n + 1):
            if i*i > n:
                break
            win[i * i] = True
        
        
        for i in range(1, n + 1):
            j = 1
            #if i is a losing position, i + j**2 is a winning position for any j>=1
            if win[i] == False:
                while(i + j * j <= n):
                    win[i + j * j] = True
                    j += 1
        
        #To see why this works, let k be not of the form i + j**2 where i is any losing position. Then k is losing position because, k - i**2 is a winning position for any i.

        return win[n]
            
Solution().winnerSquareGame(4)
```

# 149. Max Points on a Line

Given an array of points where points[i] = [xi, yi] represents a point on the X-Y plane, return the maximum number of points that lie on the same straight line.

**Example 1.**

Input: points = [[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]

Output: 4


**Constraints:**

1 <= points.length <= 300

points[i].length == 2

-104 <= xi, yi <= 104

All the points are unique.

```python
class Solution(object):
    def maxPoints(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """
        n = len(points)
        if len(points) == 1:
            return 1
        
        h = {}
        #h is a dictionary 
        #keys will be tuples (slope, intercept)
        #values are sets of points on the given line
        
        for i in range(n):
            for j in range(n):
                if i == j:
                    continue
                xi, yi = points[i]
                xj, yj = points[j]
                if xi != xj:
                    slope = float(yi - yj) / float(xi - xj)
                    intercept = float(slope * (0 - xj) + yj)
                else:
                  #vertical lines will be stored with slope infinity and intercept the x-value
                    slope = float('inf')
                    intercept = xj
                if (slope, intercept) in h:
                    h[(slope, intercept)].add((xi, yi))
                    h[(slope, intercept)].add((xj, yj))
                    
                else:
                    h[(slope, intercept)]={(xi, yi), (xj, yj)}
                
        maximum = 0
  
        for d in h:
            if len(h[d]) > maximum:
                maximum = len(h[d])
        return maximum
Solution().maxPoints([[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]])
```

# 126. Word Ladder II


A transformation sequence from word beginWord to word endWord using a dictionary wordList is a sequence of words 

`beginWord -> s1 -> s2 -> ... -> sk `

such that:

* Every adjacent pair of words differs by a single letter.
* Every si for 1 <= i <= k is in wordList. Note that beginWord does not need to be in wordList.
* `sk == endWord`

Given two words, beginWord and endWord, and a dictionary wordList, return all the shortest transformation sequences from beginWord to endWord, or an empty list if no such sequence exists. 

Each sequence should be returned as a list of the words [beginWord, s1, s2, ..., sk].

 

**Example 1:**

Input: beginWord = "hit", endWord = "cog", wordList = ["hot","dot","dog","lot","log","cog"]

Output: [["hit","hot","dot","dog","cog"],["hit","hot","lot","log","cog"]]

Explanation: There are 2 shortest transformation sequences:

"hit" -> "hot" -> "dot" -> "dog" -> "cog"

"hit" -> "hot" -> "lot" -> "log" -> "cog"

**Example 2:**

Input: beginWord = "hit", endWord = "cog", wordList = ["hot","dot","dog","lot","log"]

Output: []

Explanation: The endWord "cog" is not in wordList, therefore there is no valid transformation sequence.
 

**Constraints:**

1 <= beginWord.length <= 5

endWord.length == beginWord.length

1 <= wordList.length <= 1000

wordList[i].length == beginWord.length

beginWord, endWord, and wordList[i] consist of lowercase English letters.

beginWord != endWord

All the words in wordList are unique.

```python
import collections
import string

class Solution:
    def findLadders(self, beginWord: str, endWord: str, wordList):
        
        #letters = [chr(c) for c in range(ord('a'),ord('z')+1) ]
        #alphabet:
        letters = string.ascii_lowercase



        #We first design a graph implemented as a dictionary
        #nodes are words
        #d[word] is a set of neighbors of word (edges)
        d = {}
        #initialize d[word] to empty set for all nodes
        for word in wordList:
            d[word] = set()
        
        #in case beginWord is not in wordList
        d[beginWord] = set()

        for word in wordList:
          #for each word we build all possible neighbors in newword and check if the newword is a node
            for i in range(len(word)):
                for letter in letters:
                    if letter != word[i]:
                        newword = word[:i]+letter+word[i+1:]
                        if newword in d:
                            d[word].add(newword)
                        if newword == beginWord:
                            d[beginWord].add(word)
        
        
        if endWord not in d:
            return []

        q=collections.deque()
        #q is a queue
        #elements of q are [current_node, length, path]
        #length is the length of the path
        #path is a tuple
        #both length and path include the current node


        #initialize the queue:
        path = [beginWord]
        q.append([beginWord, 1, tuple(path)])

        #visited keeps track of the nodes we visited
        #since we are using bfs (breadth first search) algorighm we can ignore the visited
        # nodes because we have already found the shortest path to it
        visited = {}
        

        #initialize to false
        for word in wordList:
            visited[word] = False

        sol = []
        
        #l will be the length of the shortest path
        l = None


        while q:
            
            word, k, path = q.pop()

            #convert path to list since we are going to append a node
            path = list(path)

            #if we reached the end word and current length is more than l, we continue to the next element in the queue
            if l != None and l < k:
                continue

            #if current word is the endWord, we found the shortest path    
            if word == endWord:
                  l = k
                  sol += [path]

            #mark the node visited
            visited[word] = True

            #add neighbors (not visited) to the queue
            for neigh in d[word]:
                if visited[neigh] == False:
                    q.appendleft([neigh, k + 1, tuple(path + [neigh])])
        return sol

Solution().findLadders("hit", "cog", ["hot","dot","dog","lot","log","cog"])
```
